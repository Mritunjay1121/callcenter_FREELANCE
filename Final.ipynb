```python
dist1=new_df[new_df['COUNCIL DISTRICT']==1]
dist1
```

```python
print(len(list(dist1['NEIGHBORHOOD'].value_counts().index)))
dist1['NEIGHBORHOOD'].value_counts()
```

```python
dist1[dist1['NEIGHBORHOOD']=="Shoal Creek"]['TYPE'].value_counts()
```

```python
d1=dict(dist1['TYPE'].value_counts())
d1
```

```python
len(d1)
```

```python
dist1[dist1['TYPE']=='Street']['REQUEST TYPE'].value_counts()
```

```python
dist1['CREATION DATE']=pd.to_datetime(dist1['CREATION DATE'])
dist1['CLOSED DATE']=pd.to_datetime(dist1['CLOSED DATE'])
dist1['days_difference'] = (dist1['CLOSED DATE'] - dist1['CREATION DATE']).dt.days
```

```python
mx=max(dist1['days_difference'])
print(mx)
most_time_taking_problem=dist1[dist1['days_difference']==mx]['TYPE']
most_time_taking_problem
```

```python
dist1[dist1['days_difference']==mx]['NEIGHBORHOOD']
```
